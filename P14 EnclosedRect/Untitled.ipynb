```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

teto = cv2.imread('teto.jpg')
teto_gr = cv2.cvtColor(teto, cv2.COLOR_BGR2GRAY)
_, teto_th = cv2.threshold(teto_gr, 10, 255, 0)
plt.imshow(teto_th)
```

```python
# Decresing Noise
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
teto_th = cv2.morphologyEx(teto_th,cv2.MORPH_CLOSE, kernel)
plt.imshow(teto_th)
```

```python
contour, _ = cv2.findContours(teto_th, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
n_contour_l = len(contour)

color_lines = plt.get_cmap('rainbow')(np.arange(n_contour_l)/(n_contour_l-1))[:,[2,1,0]]*255

cnt_dp1 = [cv2.approxPolyDP(cnt,5,True) for cnt in contour]
cnt_dp2 = [cv2.approxPolyDP(cnt,11,True) for cnt in contour]

teto_cnt = teto.copy()
teto_cnt_dp1 = teto.copy()
teto_cnt_dp2 = teto.copy()
for i in range(n_contour_l):
    teto_cnt = cv2.drawContours(teto_cnt,contour,i,color_lines[i],2)
    teto_cnt_dp1 = cv2.drawContours(teto_cnt_dp1,cnt_dp1,i,color_lines[i],2)
    teto_cnt_dp2 = cv2.drawContours(teto_cnt_dp2,cnt_dp2,i,color_lines[i],2)

tetos = [teto_cnt, teto_cnt_dp1, teto_cnt_dp2]

plt.figure(figsize=[20,18])
for i,tet in enumerate(tetos):
    plt.subplot(321+i)
    plt.imshow(tet)

plt.tight_layout()
plt.show()
```

```python
teto_cnt = teto.copy()
for i in range(n_contour_l):
    cnt = cv2.convexHull(contour[i])
    teto_cnt = cv2.drawContours(teto_cnt, [cnt], 0 ,color_lines[i], 5)
plt.imshow(teto_cnt[:,:,::-1])
```

```python
rin = cv2.imread('rin.jpg')
rin_gr = cv2.cvtColor(rin, cv2.COLOR_BGR2GRAY)
_, rin_th = cv2.threshold(rin_gr,10,255,0)

kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
rin_th = cv2.morphologyEx(rin_th,cv2.MORPH_CLOSE, kernel)

contour,_ = cv2.findContours(rin_th,cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
n_contour_l = len(contour)

si = plt.get_cmap('rainbow')(np.arange(n_contour_l)/(n_contour_l-1))[:,[2,1,0]]*255
```

```python
rin_cnt = rin.copy()
rect = []
for i,cnt in enumerate(contour):
    rin_cnt = cv2.drawContours(rin_cnt, [cnt],0,si[i],2)
    rect_ = cv2.minAreaRect(cnt)
    rect.append(rect_)
    
for i in range(n_contour_l):
    box = cv2.boxPoints(rect[i]).astype(int)
    rin_cnt = cv2.drawContours(rin_cnt,[box],0,si[i],2)

    plt.imshow(rin_cnt[:,:,::-1])
```
