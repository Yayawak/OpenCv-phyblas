```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

rin = cv2.imread('rin.jpg')
_,thresh = cv2.threshold(
    cv2.cvtColor(rin,cv2.COLOR_BGR2GRAY),10
    ,255, 0)
plt.imshow(thresh)
```

```python
n,lek = cv2.connectedComponents(thresh)
print(lek)
```

```python
plt.imshow(lek)
```

```python
plt.imshow(lek,cmap='rainbow')
plt.colorbar(pad=.01)
plt.show()
```

```python
gumi = cv2.imread('gumi.jpg')
_,thresh = cv2.threshold(
    cv2.cvtColor(gumi,cv2.COLOR_RGB2GRAY),
    10,255,0)
n,lek,stat,centroid = cv2.connectedComponentsWithStats(thresh)
print(stat)
```

```python
plt.imshow(lek,cmap='rainbow')
plt.colorbar(pad=.01)
plt.scatter(x=centroid[:,0], y=centroid[:,1],c='k')
plt.show()
```

```python
print(centroid[:,:])
```

```python
print(centroid[:,0]) # show x-element
```

```python
print(centroid[0,:]) # show first-coord x,y
```

```python
ax = plt.axes()
plt.imshow(lek,cmap='rainbow')
for (x,y,w,h,a),(cx,cy) in zip(stat,centroid):
    ax.add_patch(plt.Rectangle((x,y),w,h,fill=False, ec='k',lw=2,ls=':'))
    plt.text(cx,cy,a,ha='center',color='k')
```
