```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt

miku = cv2.imread('miku.png')
mk = cv2.cvtColor(miku, cv2.COLOR_BGRA2RGBA)
plt.imshow(mk)
plt.show
```

```python
rin = cv2.imread('rin.jpg')
rin_gray = cv2.cvtColor(rin, cv2.COLOR_BGR2GRAY)
cv2.imwrite('rin_gray2.jpg', rin_gray)
```

```python
hsv = np.zeros([256,256,3],dtype=np.uint8)
s,v = k = np.meshgrid(np.arange(256),np.arange(256))

# print(k)
# print(s)
# print()
# print(v)

hsv[:,:,0] = 150
hsv[:,:,1] = s
hsv[:,:,2] = v

rup = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
plt.gca(title='h = 150',xlabel='s',ylabel='v')
plt.imshow(rup)
plt.show()
```
