```python
import cv2
import matplotlib.pyplot as plt

rin = cv2.imread('rin.jpg')
rinDown = cv2.resize(rin, (120,90))
rinUp = cv2.resize(rinDown, (600,450), interpolation=cv2.INTER_NEAREST)
cv2.imwrite('rinUP1.jpg',rinUp)
```

```python
rinDown = cv2.resize(rin, (40,30))
rinUp = cv2.resize(rinDown, (600,450), interpolation=cv2.INTER_NEAREST)
cv2.imwrite('rinUP2.jpg',rinUp)
```

```python
plt.imshow(rin[:,:,::-1])
plt.show()
```

```python
coffeeCup = cv2.resize(rin[320:440,300:500], (15,9))
rin[320:440,300:500] = cv2.resize(coffeeCup, (200,120), interpolation=5)
cv2.imwrite('rinBlurCoffee.jpg',rin)
```
