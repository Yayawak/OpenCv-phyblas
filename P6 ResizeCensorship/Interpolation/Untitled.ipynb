```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

teto = cv2.imread('teto.jpg')
flag = [None,cv2.INTER_NEAREST,cv2.INTER_LINEAR,cv2.INTER_AREA,cv2.INTER_CUBIC,cv2.INTER_LANCZOS4]
interp = ['','nearest','linear','area','cubic','lanczos4']

plt.figure(figsize=[10,10])
for i in range(6):
    plt.subplot(321+i, title=interp[i])
    if(flag[i] != None):
        plt.imshow(cv2.resize(teto[:,:,::-1], (125,75), interpolation=flag[i]))
    else:
        plt.imshow(teto[:,:,::-1])
plt.tight_layout()
plt.show()
```

```python
teto_head = cv2.imread('teto_head.jpg')
plt.figure(figsize=[7,10])
for i in range(6):
    plt.subplot(321+i, title=interp[i])
    if(flag[i] != None):
        plt.imshow(cv2.resize(teto_head[:,:,::-1], (500,500), interpolation=flag[i]))
    else:
        plt.imshow(teto_head[:,:,::-1])
plt.tight_layout()
plt.show()
```
