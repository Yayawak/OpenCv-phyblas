```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

rin = cv2.imread('rin.jpg')
kernel = np.ones([50,50],dtype=np.float32)/25
kernel /= kernel.sum()
rin = cv2.filter2D(rin, -1, kernel)
# cv2.imwrite('rin_filtered.jpg', rin)

plt.imshow(rin[:,:,::-1])
```

```python
rin = cv2.imread('rin.jpg')
rin_blur = cv2.blur(rin, (100,4))
plt.imshow(rin_blur[:,:,::-1],)
```
